```python
# module
import numpy as np
import pandas as pd
```

```python
# データ読み込み
train = pd.read_csv('/data/input/train.csv')
test = pd.read_csv('/data/input/test.csv')
```

```python
pd.set_option('display.max_columns', 26)
train.head(10)
```

```python
train.corr()
```

## カラム削除
- groupId
強い人と同groupIdの場合、勝率に関与すると考えられるが今回は削除。
- matchId
勝率とは無縁のため削除。

```python
def delete_column(data):
  data = data.drop(['groupId','matchId'], axis=1)
  return data

train = delete_column(train)
test = delete_column(test)
```

説明変数と目的変数の抽出/分離

```python
# trainデータのId削除
train = train.drop('Id', axis=1)

# testデータのIdを抽出し削除
test_id = test['Id']
test = test.drop('Id',axis=1)
```

```python
# 訓練用データ作成
X = train.drop('winPlacePerc',axis=1)
Y = train['winPlacePerc']
```

```python
print(X.shape)
print(test.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=2, random_state=0)
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
# 平均二乗誤差 (MSE)
from sklearn.metrics import mean_squared_error

# xgboost
model = xgb.XGBRegressor()

# ハイパーパラメータ探索
reg_cv = GridSearchCV(model, {'max_depth': [8], 'colsample_bytree': [1], 'subsample': [1],'n_estimators': [10], 'learning_rate':[0.13]}, verbose=1)
reg_cv.fit(X_train, y_train)
print(reg_cv.best_params_, reg_cv.best_score_)

# 改めて最適パラメータで学習
reg = xgb.XGBRegressor(**reg_cv.best_params_)
reg.fit(X_train, y_train)

# 目的変数予測結果
pred_train = reg.predict(X_train)
pred_test = reg.predict(X_test)
# 平均二乗誤差 (MSE)
print(mean_squared_error(y_train, pred_train))
print(mean_squared_error(y_test, pred_test))
```

```python
# ハイパーパラメータを最適化出来たら学習してpredict
reg = xgb.XGBRegressor(max_depth=8, n_estimators=80, learning_rate=0.9,subsample=0.8)
reg.fit(X, Y)

y_pred = reg.predict(test)
```
